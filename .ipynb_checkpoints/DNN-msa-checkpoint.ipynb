```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
import scipy.io as sio
from math import sqrt
import time
import random
from sklearn.utils import shuffle
import ipywidgets as widgets
from IPython.display import display, clear_output
from sklearn import preprocessing
np.seterr(all='raise')
```

```python
def sigmoid(x):
    return 1./(1+np.exp(-x))
def softmax(x):
    exp=np.exp(x-np.max(x,axis=1,keepdims=True))
    return exp/np.sum(exp,axis=1,keepdims=True)
def read_alpha_digit(path,subset):
    mat=np.take(sio.loadmat(path)['dat'],subset,axis=0)
    return np.array(list(map(lambda x: x.flatten(),mat.flatten())))
```

```python
class RBM:
    """ This class implements the Restricted Boltzman Machine 
    
    Attributes:
    -----------
                p: Int
                    Number of visible units
                q: Int
                    Number of hidden units
                a: numpy.array
                    Representation of the visible units
                b: numpy.array
                    Representation of the hidden units
                W: numpy.array
                    Weights of our model
    Methods:
    --------
                compute_hidden(X)
                    Compute the probabilities of the hidden layer
                compute_visible(Y)
                    Compute the probabilies of the visible layer
                train_one_iter(X,batch_size,lr)
                    Training for one iteration our model
                train(X,nb_iter,batch_size,lr,plot_error=False,plot_start=0)
                    Training our model for the dataset X
                generate_data(nb_iter_gibbs,nb_images)
                    Generate images from the learned distribution
                
    """
    def __init__(self,p,q):
        self.p=p
        self.q=q
        self.a=np.zeros((1,p))
        self.b=np.zeros((1,q))
        self.W=np.random.normal(0,0.1,(p,q))
        
    def compute_hidden(self,X):
        """ Compute the probabilies of the visible layer
        
        Parameters:
        -----------
            X: numpy.array
                The vector from which we compute
        
        Returns:
        --------
            numpy.array
                The probabilities of the hidden layer
        """
        return(sigmoid(self.b+X.dot(self.W)))
    
    def compute_visible(self,Y):
        """ Compute the probabilies of the hidden layer
        
        Parameters:
        -----------
            Y: numpy.array
                The vector from which we compute
        
        Returns:
        --------
            numpy.array
                The probabilities of the visible layer
        """
        return(sigmoid(self.a+Y.dot(self.W.T)))
    
    def train_one_iter(self,X,batch_size,lr):
        """ Training for one iteration our model
        
        Parameters:
        -----------
            X: numpy.array
                The dataset we want to learn the distribution
            batch_size: int
                The size of the mini-batch
            lr: float
                The learning rate for the gradient
        """
        n,p=X.shape
        shuffled_X=np.random.permutation(X)
        nb_mini_batch=n//batch_size
        for i in range(0,nb_mini_batch):
            batch=shuffled_X[i*batch_size:min(n-1,(i+1)*batch_size),]
            v0=batch 
            p_h_v0=self.compute_hidden(v0)
            h0=np.where(p_h_v0>=np.random.uniform(size=p_h_v0.shape),1,0)
            p_v_h0=self.compute_visible(h0)
            v1=np.where(p_v_h0>=np.random.uniform(size=p_v_h0.shape),1,0)
            p_h_v1=self.compute_hidden(v1)
            da=(v0-v1).sum(axis=0)
            db=(p_h_v0-p_h_v1).sum(axis=0)
            dW=v0.T.dot(p_h_v0)-v1.T.dot(p_h_v1)
            self.a=self.a+(lr/batch_size)*da
            self.b=self.b+(lr/batch_size)*db
            self.W=self.W+(lr/batch_size)*dW
            
    def train(self,X,nb_iter_grad,batch_size,lr,plot_error=False,plot_start=0):
        """ Training our model for the dataset X
        
        Parameters:
        -----------
            X: numpy.array
                The dataset we want to learn the distribution
            nb_iter_grad: int
                Number of itÃ©ration for gradient descent
            batch_size: int
                The size of the mini-batch
            lr: float
                The learning rate for the gradient
            plot_error: bool (optional)
                Boolean for ploting the error of our RBM (default = False)
            plot_start: int (optional)
                Iteration where the plot starts
        """
        if(plot_error):
            erreur_quadra=[]
            train_rbm_output=widgets.Output()
            display(train_rbm_output)
            for i in range(plot_start):
                train_rbm_output.clear_output(wait=True)
                with train_rbm_output:
                    print('RBM training. Epoch number '+str(i+1))
                self.train_one_iter(X,batch_size,lr)
            erreur_quadra=[]
            epochs=[]
            for i in range(plot_start,nb_iter_grad):
                self.train_one_iter(X,batch_size,lr)
                sortie=self.compute_hidden(X)
                nouvelle_entree=self.compute_visible(sortie)
                erreur_quadra+=[sqrt(((nouvelle_entree-X)**2).sum())/(X.shape[0]*X.shape[1])]
                epochs+=[i]
                train_rbm_output.clear_output(wait=True)
                with train_rbm_output:
                    print('RBM training. Epoch number '+str(i+1))
                    plt.plot(epochs,erreur_quadra,label='q = '+str(self.q))
            plot_range=np.linspace(plot_start,nb_iter_grad,num=nb_iter_grad-plot_start)
            plt.plot(plot_range,erreur_quadra[-nb_iter_grad+plot_start:],label='q = '+str(self.q))
        else:
            train_rbm_output=widgets.Output()
            display(train_rbm_output)
            for i in range(nb_iter_grad):
                train_rbm_output.clear_output(wait=True)
                with train_rbm_output:
                    print('RBM training. Epoch number '+str(i+1))
                self.train_one_iter(X,batch_size,lr)
            
    def generate_data(self,nb_iter_gibbs,nb_image):
        """ Generate images from the learned distribution
        
        Parameters:
        -----------
            nb_iter_gibbs: int
                Number of iteration for Gibbs Sampling
            nb_image: int
                Number of images we want to generate
        
        Returns:
        --------
            numpy.array
                The images generated with Gibbs Sampling
        """
        images=[]
        for i in range(nb_image):
            v=np.random.normal(0,0.1,(1,self.p))
            p_h_v=self.compute_hidden(v)
            for i in range(nb_iter_gibbs):
                h=np.where(p_h_v>=np.random.uniform(size=p_h_v.shape),1,0)
                p_v_h=self.compute_visible(h)
                v=np.where(p_v_h>=np.random.uniform(size=p_v_h.shape),1,0)
                p_h_v=self.compute_hidden(v)
            images+=[v]
        return(images)
```

```python
class DBN:
    """ This class implements the Deep Belief Network
    
    Attributes:
    -----------
                dims: np.array
                    Dimension of the layers of the DBN
                layers: np.array
                    Array of the RBMs of the DBN
    Methods:
    --------
                train(X,nb_iter,batch_size,lr,plot_error=False)
                    Training our model for the dataset X
                generate_data(nb_iter_gibbs,nb_images)
                    Generate images from the learned distribution
                
    """
    def __init__(self,dims):
        self.dims=dims
        self.layers=[]
        for i in range(len(dims)-1):
            self.layers+=[RBM(dims[i],dims[i+1])]
    def train(self,X,nb_iter_grad,batch_size,lr,plot_error=False):
        """ Training our model for the dataset X
        
        Parameters:
        -----------
            X: numpy.array
                The dataset we want to learn the distribution
            nb_iter_grad: int
                Number of itÃ©ration for gradient descent
            batch_size: int
                The size of the mini-batch
            lr: float
                The learning rate for the gradient
            plot_error: bool (optional)
                Boolean for ploting the error of our RBM (default = False)
        """
        v=X
        for RBM in self.layers:
            RBM.train(v,nb_iter_grad,batch_size,lr,plot_error=plot_error)
            v=RBM.compute_hidden(v)
    def generate_data(self,iter_gibbs,nb_image):
        """ Generate images from the learned distribution
        
        Parameters:
        -----------
            nb_iter_gibbs: int
                Number of iteration for Gibbs Sampling
            nb_image: int
                Number of images we want to generate
        
        Returns:
        --------
            numpy.array
                The images generated with Gibbs Sampling
        """
        RBM=self.layers[-1]
        images=RBM.generate_data(iter_gibbs,nb_image)
        for RBM in list(reversed(self.layers))[1:]:
            images_bis=[]
            for image in images:
                p_v_h=RBM.compute_visible(image)
                images_bis+=[np.where(p_v_h>=np.random.uniform(size=p_v_h.shape),1,0)]
            images=images_bis
        return(images)
```

```python
class DNN_layer:
    """ This represent one neural network layer
    
    Parameters:
    -----------
        nb_prec: int
            This parameter indicate the number of neurons in the last layer. 
            In the case of the first layer, it's the data dimensions

        nb_neurons: int
            This parmeter indicate the number of neurons in the current layers

    Methods:
    --------
    forward_propagation(X)
        This function compute the output of the layer of the neural network
        
    init_from_weights_and_bias(W,b)
        Init our layer with given weight and bias
        
    back_propagation(grad_W,grad_b,lr)
        This function compute the update to the weights and bias of the layer
    """
    
    def __init__(self,nb_prec=1,nb_neurons=1):
        self.W=np.random.randn(nb_prec,nb_neurons)*0.01
        self.b=np.zeros((1,nb_neurons))
    def forward_propagation(self,X):
        """ This function compute the output of the layer of the neural network
        
        Parameters:
        -----------
            X: numpy.array
                The data, or the precedent output
        Returns:
        --------
            np.array
                ACtivation of the layer
        
        """
        return(sigmoid(X.dot(self.W)+self.b))
    def init_from_weights_and_bias(self,W,b):
        """ This function allow us to initialize our layer from a given weight and bias 
        
        Parameters:
        -----------
            W: numpy.array
                The weight matrix we want for our layer
            b: numpy.array
                The bias vector we want for our layer
        """
        self.W=W
        self.b=b
    def back_propagation(self,grad_W,grad_b,lr):
        """ This function compute the update to the weights and bias of the layer 
        
        Parameters:
        -----------
            grad_W: numpy.array
                Computed gradient for the weights of the layer
            grad_b: numpy.array
                Computed gradient for the biases of the layer
            lr: float
                Learning rate for gradient descent
        """
        self.W=self.W-lr*grad_W
        self.b=self.b-lr*grad_b
```

```python
class DNN:
    """ This represent a neural network
    
    Attributes:
    -----------
        dims: int
            This parameter indicate the number of neurons in each layer. 
    

    Methods:
    --------
        init_from_DBN(DBN)
            This function allow for initialize our neural network from a trained DBN
            
        forward_propagation(X)
            This function compute the output of the neural network for a given data-set
            
        compute_loss(X,y)
            This function compute cross-entropy loss of the given data-set
                         
        error_rate(X,y)
            This function compute the error rate of the given data-set
                         
        compute_metrics(x,y)
            This function compute cross-entropy loss and the error rate of the given data-set

        back_propagation_step(X_train,y_train,lr,batch_size)
            This function compute one step of back-propagation

        back_propagation(X_train,y_train,lr,epochs,batch_size)
            This function compute back-propagation to train the neural network
                         
        back_propagation_with_metrics(X_train,y_train,lr,epochs,batch_size,X_test,y_test)
            This function train our neural network via back-propagation and output the loss and the accuracy for each epoch

                
    """
        
        
    def __init__(self,dims):
        self.dims=dims
        self.layers=[]
        for i in range(len(dims)-1):
            self.layers+=[DNN_layer(dims[i],dims[i+1])]
            
    def init_from_dbn(self,DBN,output_dim):
        """ This function allow for initialize our neural network from a trained DBN 
        
        Parameters:
        -----------
            DBN: DBN 
                A DBN from wich we want to initialize our network
        
        """
        self.layers=[]
        for rbm in DBN.layers:
            new_layer=DNN_layer()
            new_layer.init_from_weights_and_bias(rbm.W,rbm.b)
            self.layers+=[new_layer]
        self.layers+=[DNN_layer(DBN.layers[-1].q,output_dim)]
            
    def forward_propagation(self,X):
        """ This function compute the output of the neural network for a given data-set 
        
        Parameters:
        -----------
            X: the data
        
         Returns:    
         --------
             np.array
             The output of the neural by the sigmoid function except for
             the last layer which use a softmax function
        
        
        """
        prec=X
        for i in range(len(self.layers)-1):
            prec=self.layers[i].forward_propagation(prec)
        return softmax(prec.dot(self.layers[-1].W)+self.layers[-1].b)
    
    def compute_loss(self,X,y):
        """ This function compute cross-entropy loss of the given data-set
        
            Parameters:
            -----------
                X: numpy.array
                    A dataset we want to evaluate
                y: numpy.array
                    The truth vector of our dataset
                    
            Returns:
            --------
                 float
                     Cross entropy of the prediction of X and the truth vector
        
        """
        n,k=X.shape
        y_hat=self.forward_propagation(X)
        return(-(y*np.log(y_hat)).sum()/n)
    
    def error_rate(self,X,y):
        """ This function compute the error rate of the given data-set
        
         Parameters:
         -----------
            X: numpy.array
                A dataset we want to evaluate
            y: numpy.array
                The truth vector ofour dataset
        
        Returns:
        --------
            float
                The error rate. We consider that the classification value is the max of the probabilities
        
        """
        y_hat=self.forward_propagation(X)
        row_maxes = y_hat.max(axis=1,keepdims=True)
        y_hat_max=np.where(y_hat == row_maxes, 1, 0)
        return(((np.abs(y-y_hat_max)/2).sum(axis=1)).mean())
    
    def compute_metrics(self,X,y):
        """ This function compute cross-entropy loss and the error rate of the given data-set 
        
         Parameters:
         -----------
            X: numpy.array
                A dataset we want to evaluate
            y: numpy.array
                The truth vector ofour dataset
        
        Returns:
        --------
            float
                Cross entropy of the prediction of X and the truth vector
            float
                The error rate. We consider that the classification value is the max of the probabilities
        
        """
        n,k=X.shape
        y_hat=self.forward_propagation(X)
        row_maxes = y_hat.max(axis=1,keepdims=True)
        y_hat_max=np.where(y_hat == row_maxes, 1, 0)
        cross_entropy=-(y*np.log(y_hat)).sum()/n
        accuracy=((np.abs(y-y_hat_max)/2).sum(axis=1)).mean()
        return(cross_entropy,accuracy)
    
    def back_propagation_step(self,X,y,lr,batch_size):
        """ This function compute one step of back-propagation
        
        Parameters:
        -----------
            X_train: numpy.array
                The data we want to train our NN on.
            y_train: numpy.array
                The truth value of our data
            lr: float
                The learning rate of our gradient
            batch_size: int
                The size of the minibatch        
        
        """
        lr=lr/batch_size
        n,p=X.shape
        cross_ent=[]
        shuffled_X,shuffled_y=shuffle(X,y)
        nb_mini_batch=n//batch_size
        for i in range(0,nb_mini_batch):
            batch_X,batch_y=shuffled_X[i*batch_size:min(n,(i+1)*batch_size),],shuffled_y[i*batch_size:min(n,(i+1)*batch_size),]
            output=[batch_X]
            prec=batch_X
            for j in range(len(self.layers)-1):
                prec=self.layers[j].forward_propagation(prec)
                output+=[prec]
            batch_y_hat=softmax(prec.dot(self.layers[-1].W)+self.layers[-1].b)
            output+=[batch_y_hat]
            C=batch_y_hat-batch_y
            dW=output[-2].T.dot(C)
            db=C.sum(axis=0,keepdims=True)
            for k in range(1,len(self.layers)):
                dw_old=dW
                db_old=db
                C=(output[-k-1]*(1-output[-k-1]))*C.dot(self.layers[-k].W.T)
                dW=output[-k-2].T.dot(C)
                db=C.sum(axis=0,keepdims=True)
                self.layers[-k].back_propagation(dw_old,db_old,lr)
            self.layers[-len(self.layers)].back_propagation(dW,db,lr)
            
    def back_propagation(self,X_train,y_train,lr,epochs,batch_size):
        """ This function compute back-propagation to train the neural network
        
        Parameters:
        ----------
            X_train: numpy.array
                The data we want to train our NN on.
            y_train: numpy.array
                The truth value of our data
            lr: float
                The learning rate of our gradient
            epochs: int
                The number of iterations of our backpropagation
            batch_size: int
                The size of the minibatch
        
        """
        back_prop_epochs=widgets.Output()
        display(back_prop_epochs)
        for i in range(epochs):
            back_prop_epochs.clear_output(wait=True)
            with back_prop_epochs:
                print('DNN training. Epoch number '+str(i+1))
            self.back_propagation_step(X_train,y_train,lr,batch_size)
            
    def back_propagation_with_metrics(self,X_train,y_train,lr,epochs,batch_size,X_test,y_test):
        """ This function train our neural network via back-propagation and output the loss and the accuracy for each epoch
        
        Parameters:    
            X_train: numpy.array
                The data we want to train our NN on.
            y_train: numpy.array
                The truth value of our data. One hot encoded
            lr: numpy.array
                The learning rate of our gradient
            epochs: int
                The number of iterations of our backpropagation
            batch_size: int
                The size of the minibatch
            X_test: nupy.array
                The test set we want to compute the metrics on
            y_test: numpy.array
                The truth vector associated. One hot encoded
        
        """
        cross_ent_train=[]
        cross_ent_test=[]
        error_train=[]
        error_test=[]
        back_prop_epochs=widgets.Output()
        display(back_prop_epochs)
        for i in range(epochs):
            back_prop_epochs.clear_output(wait=True)
            with back_prop_epochs:
                self.back_propagation_step(X_train,y_train,lr,batch_size)
                print('DNN training. Epoch number '+str(i+1))
            cross_entropy_train,accuracy_train=self.compute_metrics(X_train,y_train)
            cross_entropy_test,accuracy_test=self.compute_metrics(X_test,y_test)
            cross_ent_train+=[cross_entropy_train]
            cross_ent_test+=[cross_entropy_test]
            error_train+=[accuracy_train]
            error_test+=[accuracy_test]
        plt.plot(cross_ent_test,label='Test set')
        plt.ylabel('Entropie croisÃ©e normalisÃ©e')
        plt.xlabel('Epoch')
        plt.grid()
        plt.plot(cross_ent_train,label='Train set')
        plt.legend()
        plt.show()

        plt.ylabel('Taux d\'erreur')
        plt.xlabel('Epoch')
        plt.plot(error_test,label='Test set')
        plt.plot(error_train,label='Train set')
        plt.grid()
        plt.legend()
        plt.show()
        return(cross_ent_test,cross_ent_train,error_train,error_test)
```

```python
def get_one_hot(targets, nb_classes):
    res = np.eye(nb_classes)[np.array(targets).reshape(-1)]
    return res.reshape(list(targets.shape)+[nb_classes])
def loadMNIST(prefix, folder):
    intType = np.dtype( 'int32' ).newbyteorder( '>' )
    nMetaDataBytes = 4 * intType.itemsize

    data = np.fromfile( folder + "/" + prefix + '-images.idx3-ubyte', dtype = 'ubyte' )
    magicBytes, nImages, width, height = np.frombuffer( data[:nMetaDataBytes].tobytes(), intType )
    data = data[nMetaDataBytes:].astype( dtype = 'float64' ).reshape( [ nImages, width*height ] )

    labels = get_one_hot(np.fromfile( folder + "/" + prefix + '-labels.idx1-ubyte',
                          dtype = 'ubyte' )[2 * intType.itemsize:],10).astype( dtype = 'float64' )
    data[data < 128] = 0    # Black
    data[data >= 128] = 1 
    return data, labels

trainingImages, trainingLabels = loadMNIST( "train", "./data" )
testImages, testLabels = loadMNIST( "t10k", "./data" )
# trainingImages,testImages = preprocessing.scale(trainingImages,axis=1),preprocessing.scale(testImages,axis=1)
```
